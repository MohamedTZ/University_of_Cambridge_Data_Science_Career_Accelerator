### 5) Perform anomaly detection with a statistical method and identify possible anomalies

#### 5.1) Use the interquartile range (IQR) method to identify outliers for each feature 5.2) Create a new column (corresponding to each feature) that will indicate (in binary – 0,1) if the value of that feature is an outlier as per IQR calculations.

```python
#Loop through each feature and calculate IQR
for feature in df.columns:
  Q1 = df[feature].quantile(0.25)
  Q3 = df[feature].quantile(0.75)
  IQR = Q3 -Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  #Create binary column for each feature and use the lambda function with if
  df[feature + '_outlier'] = df[feature].apply(lambda x: 1 if x < lower_bound or x > upper_bound else 0)

df.head()
```

```python
#Number of outliers in each feature
outliers_summary = df.iloc[:, -6:].sum()
outliers_summary_df = pd.DataFrame(outliers_summary, columns=['Number of Outliers'])
outliers_summary_df ['Percentage of Outliers'] = (outliers_summary_df['Number of Outliers'] / len(df)) * 100
outliers_summary_df
```

#### 5.3) Use IQR to identify a sample as an outlier only if two or more of the features fall under an outlier category for a particular sample.

```python
#Create another column that counts the binary for each feature
df['Total_Outliers'] = df.iloc[:, -6:-1].sum(axis=1)
df.head()
```

```python
#Viewing the count of outliers
outlier_count = df['Total_Outliers'].value_counts()
outlier_count = pd.DataFrame(outlier_count, columns=['count'])
outlier_count['outlier_percentage'] = (outlier_count['count'] / len(df)) * 100
outlier_count
```

```python
#more than one outlier
one_or_more = df[df['Total_Outliers'] >= 1]
two_or_more = df[df['Total_Outliers'] >= 2]
three_or_more = df[df['Total_Outliers'] >= 3]

print("Number of samples 1 or more outliers is {}, and the percentage is {}".format(len(one_or_more), len(one_or_more) *100 / len(df)))
print("Number of samples 2 or more outliers is {}, and the percentage is {}".format(len(two_or_more), len(two_or_more) *100 / len(df)))
print("Number of samples 3 or more outliers is {}, and the percentage is {}".format(len(three_or_more), len(three_or_more) *100 / len(df)))
```

```python
#Import libraries
from sklearn.preprocessing import StandardScaler
from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.decomposition import PCA
```

#### 6.1) Perform feature scaling to prepare the data for ML algorithms.

```python
#Create instance of StandardScaler
scaler = StandardScaler()

#Only extracting relevant features from original data
df_2 = df.iloc[:, :6]

#Fit and transform the data
scaled_data = scaler.fit_transform(df_2)

#Converting scaled_data variable into a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=df_2.columns)

scaled_df.head()
```

```python
#Create instance of one clase SVM, starting with high gamma and low nu
one_class_svm = OneClassSVM(kernel='rbf', gamma=0.5, nu=0.05)

#Fit the model
one_class_svm.fit(scaled_df)
```

```python
#Predict the values (-1 and 1) from the model
y_pred = one_class_svm.predict(scaled_df)
y_pred
```

```python
#Add y_pred to scaled_df
scaled_df['y_pred'] = y_pred
scaled_df.head()
```

```python
#Subset y_pred where the value is -1
outliers_svm = scaled_df[scaled_df['y_pred'] == -1]
outliers_svm.head()
```

```python
print("Number of anomalies using One Class SVM (gamma = 0.5, nu = 0.05) is {}, and percentage is {}".format(outliers_svm.shape[0], outliers_svm.shape[0] *100 / len(df) ))
```

Given that 5.3% is outside the expected range, one class SVM model neeeds to be fine tuned since my initial arbitary values for gamma and nu are 0.5 and 0.05 respectively.
